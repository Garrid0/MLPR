```python
import numpy as np
import matplotlib.pyplot as plt

def gauss(xx, mu, var):
    return (1/np.sqrt(2*np.pi*var)) * np.exp((-(xx-mu)**2)/(2*var))

x1 = [0.5, 0.1, 0.2, 0.4, 0.3, 0.2, 0.2, 0.1, 0.35, 0.25]
mu1, var1 = np.mean(x1), np.var(x1)

x2 = [0.9, 0.8, 0.75, 1.0]
mu2, var2 = np.mean(x2), np.var(x2)

pi1, pi2 = len(x1)/(len(x1)+len(x2)), len(x2)/(len(x1)+len(x2))

xx_grid = np.linspace(0, 1, 100, endpoint=False)

prob_class1 = gauss(xx_grid, mu1, var1)*pi1
prob_class2 = gauss(xx_grid, mu2, var2)*pi2
prob_x_is_0point6 = gauss(0.6, mu1, var1)*pi1 + gauss(0.6, mu2, var2)*pi2
prob_y_is_1_given_x_is_0point6 = (gauss(0.6, mu1, var1)*pi1)/prob_x_is_0point6

plt.figure(figsize=(8,6))
plt.plot(xx_grid, prob_class1, "r", label="p(x,y=1)")
plt.plot(xx_grid, prob_class2, "b", label="p(x,y=2)")
plt.legend()
plt.xlabel("Input location, x")
plt.ylabel("Scores, p(x,y)")
plt.show()

print(f"P(y=1|x=0.6) = {round(prob_y_is_1_given_x_is_0point6,2)} 2s.f")
```
